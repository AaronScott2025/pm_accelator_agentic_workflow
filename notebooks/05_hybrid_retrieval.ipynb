# Hybrid Retrieval (Dense + BM25)

Demonstrate linear fusion reranking.

```python
from ai_interviewer_pm.retrieval.hybrid import BM25Retriever, hybrid_rerank
from ai_interviewer_pm.retrieval.vectorstore import build_vectorstore, similarity_search
corpus = ['pm interview questions','behavioral interview tips','neural embeddings intro']
bm = BM25Retriever(corpus)
sparse = bm.search('interview', k=3)
dense_store = build_vectorstore(corpus, metadatas=[{}]*len(corpus))
dense = [(t, s) for (t, _m, s) in similarity_search(dense_store, 'interview', k=3)]
hybrid = hybrid_rerank(dense, sparse, alpha=0.6)
hybrid
```
